### 유용한 아주 기초 리눅스 명령어 
- `ls`: 현재 디렉토리의 파일 및 폴더 리스트 확인
- `cd 폴더명`: "폴더명" 폴더로 이동
- `cd ..`: 상위 폴더로 이동 
- `mv 원본파일명 이동경로/파일명`: 파일을 다른 위치로 이동 또는 이름변경 
- `cp -rf 원본파일명 사본파일명`: 파일이나 하위 폴더 전체를 복사 
- `rm -rf 파일명`: 파일이나 하위폴더 전체를 삭제

### 많은 양의 이미지 데이터를 서버로 한꺼번에 업로드 하는 방법 
-  .tar로 압축하는 경우 
    - 압축파일을 서버에 업로드 
    - terminal에서 해당 경로를 찾아감
    - `tar -xvzf 압축파일명.tar` 명령어 실행 
    - 압축해제가 완료된 후 `ls` 명령어로 파일이 있는지 확인 
- .zip으로 압축하는 경우 
    - 압축파일을 서버에 업로드 
    - terminal에서 해당 경로를 찾아감
    - `unzip 압축파일명.zip` 명령어 실행 
    - 압축해제가 완료된 후 `ls` 명령어로 파일이 있는지 확인 
- 주피터 노트북에서 폴더 경로를 찾아갔을 때 파일이 정상적으로 보이지 않더라도  terminal 환경에서 확인되면 ok
